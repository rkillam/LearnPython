```python
#!/usr/bin/env python

# __future__ module helps with Python 2 to 3 conversions

# division uses Python 3's division operator which always does float division
# therefore 5 / 2 = 2.5 instead of 2
# for integer division use //, 5 // 2 = 2
from __future__ import division

# print_function uses Python 3's print function which allows us to control
# what is printed at the end using end:
#          print('hello world', end='!@#') => hello world!@#
from __future__ import print_function

import random

def get_data():
    """Return 3 random integers between 0 and 9"""
    return random.sample(range(10), 3)

def consume():
    """Displays a running average across lists of integers sent to it"""
    running_sum = 0
    data_items_seen = 0
    tot_data = []

    while True:
        data = yield
        data_items_seen += len(data)
        running_sum += sum(data)
        tot_data.append(data)
        print('The running average of {} is {}'.format(tot_data, running_sum / float(data_items_seen)))

def produce(consumer):
    """Produces a set of values and forwards them to the pre-defined consumer
    function"""
    while True:
        data = get_data()
        print('Produced {}'.format(data))
        consumer.send(data)
        yield

if __name__ == '__main__':
    consumer = consume()
    consumer.send(None)
    producer = produce(consumer)

    for _ in range(10):
        print('\nProducing...')
        next(producer)
```
