```python
#!/usr/bin/env python

# __future__ module helps with Python 2 to 3 conversions

# division uses Python 3's division operator which always does float division
# therefore 5 / 2 = 2.5 instead of 2
# for integer division use //, 5 // 2 = 2
from __future__ import division

# print_function uses Python 3's print function which allows us to control
# what is printed at the end using end:
#          print('hello world', end='!@#') => hello world!@#
from __future__ import print_function


def arg_decorator(a, b, c):
    def wrapper(f):
        print('wrapper')
        def inner(*args, **kwargs):
            print('inner: ', a, b, c)
            print(args)
            print(kwargs)

            try:
                return f(*args, **kwargs)
            except:
                print("in inner except")
            return "inner finally"

        return inner
    return wrapper


@arg_decorator('A', 'B', 'C')
def foo1(x, y=10):
    print('foo1: ', x, y)

    if x < 0:
        print("fake exception")
        raise Exception("fake exception")

    return x + y

def foo2(x, y=10):
    print('foo2: ', x, y)

    if x < 0:
        print("fake exception")
        raise Exception("fake exception")

    return x + y


# These produce the same result, but in an uglier syntax
# It gives a better understanding to what decorators are doing
wrapper = arg_decorator('A', 'B', 'C')
foo2 = wrapper(foo2)

foo3 = arg_decorator('A', 'B', 'C')(foo2)

print('=' * 40)
print('foo1')
print(foo1(1))
print()
print(foo1(1, 2))
print()
print(foo1(1, y=3))
print()
print(foo1(y=4, x=2))
print()
print(foo1(-1))
print('=' * 40)

print('=' * 40)
print('foo2')
print(foo2(1))
print()
print(foo2(1, 2))
print()
print(foo2(1, y=3))
print()
print(foo2(y=4, x=2))
print()
print(foo2(-1))
print('=' * 40)

print('=' * 40)
print('foo3')
print(foo3(1))
print()
print(foo3(1, 2))
print()
print(foo3(1, y=3))
print()
print(foo3(y=4, x=2))
print()
print(foo3(-1))
print('=' * 40)
```
