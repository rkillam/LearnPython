```python
#!/usr/bin/env python

# __future__ module helps with Python 2 to 3 conversions

# division uses Python 3's division operator which always does float division
# therefore 5 / 2 = 2.5 instead of 2
# for integer division use //, 5 // 2 = 2
from __future__ import division

# print_function uses Python 3's print function which allows us to control
# what is printed at the end using end:
#          print('hello world', end='!@#') => hello world!@#
from __future__ import print_function

# When you decorate a function/class you change its meta data (name, module, etc)
# By decorating your wrapper function with wraps, it keeps the original objects metadata
from functools import wraps


def non_neg_args(f):
    @wraps(f)
    def wrapper(*args, **kwargs):
        abs_args = []
        for arg in args:
            abs_args.append(abs(arg))

        for key, arg in kwargs.values():
            kwargs[key] = abs(arg)

        return f(*abs_args, **kwargs)
    return wrapper


def no_wraps_non_neg_args(f):
    def wrapper(*args, **kwargs):
        abs_args = []
        for arg in args:
            abs_args.append(abs(arg))

        for key, arg in kwargs.values():
            kwargs[key] = abs(arg)

        return f(*abs_args, **kwargs)
    return wrapper


@no_wraps_non_neg_args
def no_wrap_add(x, y):
    return x + y


@non_neg_args
def add(x, y):
    return x + y

print(add.__name__)
print(add.__module__)

print(no_wrap_add.__name__)
print(no_wrap_add.__module__)

print()

x, y = 1, 2
print('add({}, {}) -> {} == {}'.format(x, y, add(x, y), 3))
print('no_wrap_add({}, {}) -> {} == {}'.format(x, y, no_wrap_add(x, y), 3))

x, y = -1, 2
print('add({}, {}) -> {} == {}'.format(x, y, add(x, y), 3))
print('no_wrap_add({}, {}) -> {} == {}'.format(x, y, no_wrap_add(x, y), 3))

x, y = -1, -2
print('add({}, {}) -> {} == {}'.format(x, y, add(x, y), 3))
print('no_wrap_add({}, {}) -> {} == {}'.format(x, y, no_wrap_add(x, y), 3))
```
